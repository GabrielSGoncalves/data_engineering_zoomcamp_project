# Question 3. Count records
How many taxi trips were totally made on January 15? 

Tip: started and finished on 2019-01-15.

### Reading the NYC dataset for January of 2019

````python
df_nyc_jan19 = pd.read_csv('datasets/green_tripdata_2019-01.csv.gz', compression='gzip')
````

````python
df_nyc_jan19.lpep_pickup_datetime = pd.to_datetime(df_nyc_jan19.lpep_pickup_datetime)
df_nyc_jan19.lpep_dropoff_datetime = pd.to_datetime(df_nyc_jan19.lpep_dropoff_datetime) 
````

````python
df_nyc_jan19.dtypes
````

````python
from datetime import date
````

````python
date(2019, 1, 15)
````

````python
df_nyc_jan19[
    (df_nyc_jan19.lpep_pickup_datetime.dt.date == date(2019, 1, 15)) &
    (df_nyc_jan19.lpep_dropoff_datetime.dt.date == date(2019, 1, 15))
]
````

## Question 4. Largest trip for each day

## Question 5. The number of passengers
In 2019-01-01 how many trips had 2 and 3 passengers?

````python
df_nyc_jan19[df_nyc_jan19.lpep_pickup_datetime.dt.date == date(2019, 1, 1)].passenger_count.value_counts()
````
